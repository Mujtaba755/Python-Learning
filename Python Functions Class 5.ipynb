```python
#Lambda = short one line fucntion
cube = lambda x: x*x*x
print(cube(5))
```

```python
#Map to get squares
numbers = [1,2,3,4]
square = lambda x : x * x
result = list(map(square, numbers))
print(result)
```

```python
#filter
numbers = [1,2,3,4,5,6]
even = list(filter(lambda x: x%2==0, numbers))
print(even)
```

```python
#use filter to get num greater then 5
numbers = [1,8, 10, 23, 3]
greater5 = list(filter(lambda x: x>5, numbers))
print(greater5)
```

```python
#File Handling
with open("marks.txt", "w") as file:
    file.write("Ali's marks are 10\n")
    file.write("Aslam's marks are 20\n")
    file.write("Usman marks are 23\n")

with open("marks.txt", "a") as file:
    file.write("Abdullah's marks are 30\n")

with open("marks.txt", "r") as file:
    print(file.read())
```
