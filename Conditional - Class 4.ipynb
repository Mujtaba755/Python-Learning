```python
inpa = input("Enter First Number ")
inpb = input("Enter Second Number ")

if inpa > inpb:
    print(f"{inpa} is Greater")
elif inpa == inpb:
    print("Both Numbers are Equal")
else:
    print(f"{inpb} is Greater")
```

```python
marks = int(input("Enter Your Grade: "))

if marks >= 90:
    print("Grade A")
elif marks >= 80:
    print("Grade B")
elif marks >= 70:
    print("Grade C")
else:
    print("Fail")
```

```python
#Factorial
num = int(input("Enter Number to find its factorial: "))
factorial = 1
for i in range(1, num+1):
    factorial*=i
    print(factorial)

print(f"The Factorial of {num} is {factorial}")
```
